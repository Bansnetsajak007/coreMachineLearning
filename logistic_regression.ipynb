```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
income_data = pd.read_csv('./Product_sales.csv')

income_data.head()
income_data.shape
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(income_data['Income'], income_data['Bought'], color='red', alpha=0.5)
plt.title('Income vs Bought')
plt.xlabel('Income')
plt.ylabel('Bought')
plt.show()
```

```python
def sigmoid(z):
    """Compute the sigmoid function."""
    res = 1 / (1 + np.exp(-z)) # Round the result to 4 decimal places
    res = round(res, 4)
    return res

res = sigmoid(0.0023)
if res >= 0.5:
    print("Lies in Class 1")
else:
    print("Lies in Class 0")
```
