```python
import pandas as pd
import numpy as np
```

```python
sales_df = pd.read_csv('monthly_sales_dataset.csv')

normal_sales_df = sales_df.copy()
normal_sales_df.head()
```

```python
normal_sales_df.shape
```

```python
sales_df['half_product_price'] = sales_df['product_price'] / 2

sales_df.head()
```

```python
sales_df.shape
```

```python
x_normal = normal_sales_df.drop('monthly_sales', axis=1)
y_normal = normal_sales_df['monthly_sales']
```

```python
x_multi = sales_df.drop('monthly_sales', axis=1)
y_multi = sales_df['monthly_sales']
```

```python
from sklearn.model_selection import train_test_split
X_train_normal, X_test_normal, y_train_normal, y_test_normal = train_test_split(x_normal, y_normal, test_size=0.2, random_state=42)
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(x_multi, y_multi, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_transformed_normal = scaler.fit_transform(X_train_normal)
X_test_transformed_normal = scaler.transform(X_test_normal)

X_train_transformed_multi = scaler.fit_transform(X_train_multi)
X_test_transformed_multi = scaler.transform(X_test_multi)
```

```python
from sklearn.linear_model import LinearRegression
model_normal = LinearRegression()
model_multi = LinearRegression()
model_normal.fit(X_train_transformed_normal, y_train_normal)
model_multi.fit(X_train_transformed_multi, y_train_multi)
```

```python
y_pred_normal = model_normal.predict(X_test_transformed_normal)
y_pred_multi = model_multi.predict(X_test_transformed_multi)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np


mae_normal = mean_absolute_error(y_test_normal, y_pred_normal)
mse_normal = mean_squared_error(y_test_normal, y_pred_normal)
rmse_normal = np.sqrt(mse_normal)
r2_normal = r2_score(y_test_normal, y_pred_normal)

print(f"MAE (Normal): {mae_normal}")
print(f"MSE (Normal): {mse_normal}")
print(f"RMSE (Normal): {rmse_normal}")
print(f"R2 Score (Normal): {r2_normal}")

mae_multi = mean_absolute_error(y_test_multi, y_pred_multi)
mse_multi = mean_squared_error(y_test_multi, y_pred_multi)
rmse_multi = np.sqrt(mse_multi)
r2_multi = r2_score(y_test_multi, y_pred_multi)

print(f"MAE (Multi): {mae_multi}")
print(f"MSE (Multi): {mse_multi}")
print(f"RMSE (Multi): {rmse_multi}")
print(f"R2 Score (Multi): {r2_multi}")
```
